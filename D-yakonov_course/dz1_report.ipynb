<h1><center>Отчёт по заданию "Кластеризация"</center></h1>       

Реализация kmeans++, основанная на выборе начальных центров с учём расстояния до уже выбранных, даёт относительно малую вероятность выбора близких центров.

Алгоритм для kmeans++:
\begin{array}{l}{\text { 1a. Выбираем первый центр } c_{1}, \text { случайно из } \mathcal{X} .} \\ {\text { 1b. Выбираем новый центр } c_{i} \in \mathcal{X} \text { с вероятностью } \frac{D(x)^{2}}{\sum_{x \in \mathcal{X}} D(x)^{2}}} \\ {\text { 1c. Повторяем шаг } \text {1b. до тех пор пока не наберём } k \text { центров. }}\end{array}

Как видно это не самая лучшая задача для Kmeans. Этот алоритм хорошо определяет "шарообразные" кластеры с большой плотностью.

Kmeans также используется в кластеризации документов, выявлении криминогенных зон, сегментации клиентов, выявлении страховых мошенничеств, анализа данных общественного транспорта, кластеризации ИТ-оповещений.

 Реализовано два метода:
 \begin{array}{l}\text {1. Метод Локтя, где: }\\ \quad \quad WSS= \sum_\limits{C_{k} \in C}\sum_\limits{x_{i} \in C_{k}}\left\|x_{i}-\mu_{k}\right\|^{2} \end{array}

\begin{array}{l}\text { 2. Метод силлуэтов, где: }\\ \quad \quad a(i)=\frac{1}{|C(i)|-1} \sum_\limits{C(i), i \neq j} d(i, j) \\ \quad \quad b(i)=\min_\limits{i \neq j}\left(\frac{1}{|C(j)|} \sum_\limits{j \in C(j)} d(i, j)\right) \\ \quad \quad s(i)=\frac{b(i)-a(i)}{\max (a(i), b(i))} \\ \quad \quad 
\text {Avg silhouette = }\frac{\sum_{i=1}^n s(i)}{n}
 \end{array}



Исходя из эксперимента, при больших K "elwob_method" теряет свою предсказательную способность, в отличии от "silhouette_method"  

Время выполнения алгоритма Ллойда является ${\displaystyle O(nkdi)}$, где:

*   n - число d-мерных векторов (подлежащих кластеризации)
*   k - число кластеров
* i  - число итераций, необходимых до сходимости.
* d - размерность пространства

На данных, которые имеют кластеризованную структуру, число итераций до сходимости часто невелико, и результаты только немного улучшаются после первого десятка итераций.

В худшем случае, алгоритма Ллойда требует  ${\displaystyle i=2^{\Omega ({\sqrt {n}})}}$ итераций

Как видно упорядочивание кластеров по расстоянию от наибольшего кластера даёт дополнительную информацию об расположении кластеров. Например, два кластера достаточно близкие к друг другу образуют в совокупности больший квадрат.
